## Create Dataframe

### pd.DataFrame( )

`pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)`  

- `data` : The data you want to store in the DataFrame. It could be:
  - A dictionary of lists, arrays, or series
  - A list of dictionaries
  - A 2D NumPy array or a list of lists
  - Another DataFrame
- index: An optional parameter to specify the row labels (index). If not provided, pandas will use default integer indexing.
- columns: An optional parameter to specify the column labels. If not provided, pandas will use default column names (0, 1, 2, ...).
- dtype: You can specify the data type for all columns (e.g., float64).
- copy: If True, it copies the data; if False, it uses the original data.

````python
### Creating a DataFrame from a Dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}

df = pd.DataFrame(data)
print(df)
````

````python
### copy = True
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df_original = pd.DataFrame(data)

# Creating a new DataFrame with copy=True
df_copy = pd.DataFrame(df_original, copy=True)

# Modifying the copied DataFrame
df_copy['Age'] = df_copy['Age'] + 1

# Display both DataFrames
print("Original DataFrame:")
print(df_original)
print("\nCopied DataFrame:")
print(df_copy)


# if without copy = True
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df_original = pd.DataFrame(data)

df_reference = pd.DataFrame(df_original, copy=False)

#Modifying the referenced DataFrame
df_reference['Age'] = df_reference['Age'] + 1

#Display both DataFrames
print("Original DataFrame:")
print(df_original)
print("\nReferenced DataFrame:")
print(df_reference)

'''Pandas internally makes copies for certain operations (such as column modifications), even if copy=False was specified, to maintain internal data integrity'''
````

### df.sort_values( )

`df.sort_values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)`  

used to sort a DataFrame by the values of one or more columns.  

- `by` : str or list of str – The column or columns to sort by.
- `axis` : {0 or 'index', 1 or 'columns'} – Axis to sort along. By default, it sorts by row (axis=0).
- `ascending` : bool or list of bool – Sort in ascending (True) or descending (False) order. If multiple columns are provided, it can be a list of booleans.
- `inplace` : bool, default False – If True, it modifies the original DataFrame. If False, it returns a new DataFrame.
- `kind` : {'quicksort', 'mergesort', 'heapsort', 'stable'}, default 'quicksort' – The sorting algorithm to use.
- `na_position` : {'first', 'last'}, default 'last' – Determines whether NaNs appear at the beginning or end.
- `ignore_index` : bool, default False – If True, the resulting DataFrame will have a new integer index.
- `key` : callable, optional – A function that takes one argument and returns a transformed version of the data (used for custom sorting).


````python
data = {'Name': ['Alice', '10', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Salary': [70000,50000, 80000, 60000]}

df = pd.DataFrame(data)

# sorting by a single column
df_sorted_1 = df.sort_values(by='Salary')
# sorting by a single column with descending order
df_sorted_2 = df.sort_values(by='Salary', ascending=False)
#sorting by a single column with new index
df_sorted_3 = df.sort_values(by='Salary',ignore_index=True)
#sorting by multiple columns
df_sorted_4 = df.sort_values(by=['Salary', 'Age'])
#sorting by multiple columns with ascending and desecending order
df_sorted_5 = df.sort_values(by=['Salary', 'Age'], ascending=[True, False])
print('sorting by a single column:')
print(df_sorted_1)
print('sorting by a single column with descending order:')
print(df_sorted_2)
print('sorting by a single column with new index:')
print(df_sorted_3)
print('sorting by multiple columns:')
print(df_sorted_4)
print('sorting by multiple columns with ascending and desecending order:')
print(df_sorted_5)
````

- `level` : int, str, or tuple, optional – If the DataFrame has a MultiIndex, you can specify the level(s) to sort.
- `ascending` : bool or list of bools, default True – Whether to sort in ascending (True) or descending (False) order. It can also be a list of booleans for multiple levels (in case of a MultiIndex).
- `inplace` : bool, default False – If True, the operation is done in place, and nothing is returned.
- `kind` : {'quicksort', 'mergesort', 'heapsort', 'stable'}, default 'quicksort' – The sorting algorithm to use.
- `na_position` : {'first', 'last'}, default last – If there are missing values (NaNs), whether they should appear first or last in the sorted order.

### `pd.merge()`
`pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)`  

|Parameter	|Description	|Default|
|---|---|---|
|left	|The first DataFrame to merge.	|Required|
|right	|The second DataFrame to merge.	|Required|
|how	|Type of merge to perform: <br>`inner`:Returns rows with matching keys in both DataFrames.<br>`outer`:Returns all rows from both DataFrames, with NaN for missing matches.<br>`left`:Returns all rows from the left DataFrame, with NaN for missing matches in the right DataFrame.`right`:Returns all rows from the right DataFrame, with NaN for missing matches in the left DataFrame.	|'inner'|
|on	|Column(s) to join on. Must be present in both DataFrames.	|None|
|left_on	|Column(s) in the left DataFrame to join on.	|None|
|right_on	|Column(s) in the right DataFrame to join on.	|None|
|left_index	|Use the index from the left DataFrame as the join key.	|False|
|right_index	|Use the index from the right DataFrame as the join key.	|False|
|suffixes	|Suffixes to apply to overlapping column names in the result.	|('_x', '_y')|
|indicator	|If True, adds a column to the output indicating the source of each row ('both', 'left_only', 'right_only').	|False|
  


````python
# if a key present multiples in left and right dataframe, then the row will become left_keys * right_keys
df_left = pd.DataFrame({'key': ['A', 'B','B', 'C', 'D'],
                       'value': [1, 21, 22, 3,4]})
df_right = pd.DataFrame({'key': [ 'B', 'B','B','C','C', 'E'],
                        'value': [201,202,203,300,301, 400]})
df_merged = pd.merge(df_left, df_right, on='key')  #how='inner' is default
df_merged   #row count of B = 2*3 , row count of C = 1*2
````

````python
df_merged_left = pd.merge(df_left, df_right, on='key', how='left') #how='left'
df_merged_left
````

````python
df_merged_left = pd.merge(df_left, df_right, on='key', how='left')
df_merged_left
````

````python
df_merged_outer = pd.merge(df_left, df_right, on='key', how='outer') #how='outer'
df_merged_outer
````

### `df.groupby()`
`df.groupby(by, axis=0, level=None, as_index=True, sort=True, group_keys=True, observed=False, dropna=True)`  
|Parameters |Description|
|---|---|
|by| Column(s) or index level(s) to group by.|
|as_index|If True, the group labels are set as the index. If False, the group labels are not set as the index.|
|sort| Whether to sort the group keys.|
|dropna| Exclude groups with NaN values in the grouping key.|
   
|Method|Description|
|---|---|
|`agg()`|Apply one or more functions to the grouped data.parameter:'sum','mean','median',etc.|
|`mean()`|Calculate the mean of each group.|
|`sum()`|Calculate the sum of each group.|
|`count()`|Count the number of non-null values in each group.|
|`median()`|Calculate the median of each group.|
|`std()`/`var()`|Calculate the standard deviation/variance of each group.|
|`min()`/`max()`|Calculate the minimum/maximum value of each group.|
|`first()`/`last()`|Get the first/last value of each group.|
|`size()`|Get the size of each group.|
|`transform()`|Transform the data within each group.|
|`filter()`|Filter the data based on a condition.|
|`apply()`|Apply a function to each group and return a DataFrame.|
|`groups`|Get a dictionary mapping group names to group indices.|
|`get_group()`|Get a specific group by name.|
|`nth()`|Get the nth row of each group.|

````python
# group by a single column
data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Subcategory': ['X', 'X', 'Y', 'Y', 'X'],
        'Values1': [10, 20, 30, 40, 50],
        'Values2': [5, 15, 25, 35, 45]}

df = pd.DataFrame(data)

grouped = df.groupby('Category')
print(grouped)
````

if grouped dataframe using .apply(), x is a dataframe in the group, and can be used to apply dataframe function

````python
print(type(df.groupby('Category')))  # Check the type of the grouped object
def print_return(x):
    print(x)  # Define a function to print the group
    return x  # Return the group
df_test = df.groupby('Category').apply(print_return)  # Apply a function to each group and print the result
print(df_test)  # Display the result of the apply function
print(type(df_test))  # Check the type of the result
````

|Option	|Description|
|---|---|

|display.precision	|Number of decimal places to display for floats.|

|Option|	Description|
|---|---|
|mode.chained_assignment	|Controls whether a warning is raised for chained assignments ('warn', 'raise', or None).|
|mode.use_inf_as_na	|Treat inf and -inf as NaN when set to True.|
